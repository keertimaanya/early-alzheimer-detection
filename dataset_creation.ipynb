```python

base_path = "/content/drive/MyDrive/ADNIMERGE2/ADNIMERGE2/data/"

dx_df = list(pyreadr.read_r(base_path + "DXSUM.rda").values())[0]
demo_df = list(pyreadr.read_r(base_path + "PTDEMOG.rda").values())[0]
apoe_df = list(pyreadr.read_r(base_path + "APOERES.rda").values())[0]
```

```python
final_df = dx_df.merge(demo_df, on='RID', how='inner')
final_df = final_df.merge(apoe_df, on='RID', how='inner')
```

```python
print(demo_df.columns)
```

```python
demo_clean = demo_df[['RID', 'PTDOB', 'VISDATE', 'PTGENDER']]
```

```python
import pandas as pd

demo_clean['PTDOB'] = pd.to_datetime(demo_clean['PTDOB'], errors='coerce')
demo_clean['VISDATE'] = pd.to_datetime(demo_clean['VISDATE'], errors='coerce')
```

```python
print(demo_clean[['PTDOB', 'VISDATE']].head())
```

```python
demo_clean = demo_clean.copy()

demo_clean['AGE'] = (demo_clean['VISDATE'] - demo_clean['PTDOB']).dt.days / 365.25
```

```python
demo_clean = demo_clean[['RID', 'AGE', 'PTGENDER']]
```

```python
dx_clean = dx_df[['RID', 'VISCODE', 'DIAGNOSIS']]
```

```python
print(apoe_df.columns)
```

```python
apoe_clean = apoe_df[['RID', 'GENOTYPE']].copy()

# count number of '4' in genotype
apoe_clean['APOE4'] = apoe_clean['GENOTYPE'].str.count('4')
```

```python
apoe_clean = apoe_clean[['RID', 'APOE4']]
```

```python
final_df = dx_clean.merge(demo_clean, on='RID', how='inner')
final_df = final_df.merge(apoe_clean, on='RID', how='inner')

final_df.columns = ['RID', 'VISCODE', 'DX', 'AGE', 'PTGENDER', 'APOE4']
```

```python

merged_df = final_df.merge(
    dti_df,
    on=['RID', 'VISCODE'],
    how='inner'
)

print("Merged shape:", merged_df.shape)
```

```python
merged_df['DX'] = merged_df['DX'].replace({'Dementia': 'AD'})
```

```python
print(merged_df['DX'].value_counts())
```

```python
print(merged_df.groupby('DX')['Image Data ID'].nunique())
```

```python
# prefer PA scans
pa_df = merged_df[merged_df['Description'].str.contains('PA')]

# fallback to others if PA not available
final_df = pa_df.drop_duplicates(subset=['RID', 'VISCODE'])

# check missing patients
missing = merged_df[~merged_df['RID'].isin(final_df['RID'])]

# add one scan for missing
missing = missing.drop_duplicates(subset=['RID', 'VISCODE'])

final_df = pd.concat([final_df, missing]) #final_df updates, has processed dtidata + adnimmerge
```

```python
print(final_df.groupby(['RID', 'VISCODE']).size().max())
```

```python
final_df = final_df.reset_index(drop=True)
```

```python
mapping = {'CN': 0, 'MCI': 1, 'AD': 2}
final_df['DX'] = final_df['DX'].map(mapping)
```

```python
print(final_df['DX'].value_counts())
```

```python
final_df = final_df[
    ['RID', 'VISCODE', 'DX', 'AGE', 'APOE4', 'PTGENDER', 'Image Data ID']
]
```

```python
final_df.to_csv("final_df.csv", index=False)
```
